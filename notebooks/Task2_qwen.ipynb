# Qwen Scaled Model (Run B)

## Objective

Scale the base Qwen3-0.6B model to a larger configuration and evaluate its behavior across CPU, GPU, and TPU backends.

The assignment requires scaling the model beyond the base configuration and documenting the changes made.

````python
%cd /content/maxtext
````

````python
!sed -n '1,100p' src/maxtext/configs/models/qwen3-0.6b.yml

!sed -n '1,100p' src/maxtext/configs/models/qwen3-1.7b.yml
````

## Model Scaling Analysis

To create a larger Dense model, the existing MaxText configuration `qwen3-1.7b` was selected.

### Configuration Changes

| Parameter | Qwen3-0.6B | Qwen3-1.7B |
|------------|------------|------------|
| Embedding Dimension | 1024 | 2048 |
| MLP Dimension | 3072 | 6144 |
| Decoder Layers | 28 | 28 |
| Query Heads | 16 | 16 |
| KV Heads | 8 | 8 |
| Head Dimension | 128 | 128 |

### Why these changes increase model size

The largest increase comes from:

- Doubling the embedding dimension
- Doubling the feed-forward (MLP) dimension

These parameters dominate transformer parameter count, resulting in a much larger model while preserving the overall architecture.

### Why Qwen3-1.7B was chosen

The assignment required scaling the base Qwen3-0.6B model to approximately 1B parameters.

Qwen3-1.7B is already provided by MaxText, represents a substantial increase in capacity, and requires no custom architecture modifications.

# Qwen 1.7B - CPU Execution

````python
%%bash
cd /content/maxtext

python3 -m maxtext.trainers.pre_train.train \
src/maxtext/configs/base.yml \
model_name=qwen3-1.7b \
dataset_type=synthetic \
run_name=qwen3_1_7b_cpu \
steps=50 \
per_device_batch_size=1 \
enable_checkpointing=false
````

## CPU Result

Unlike the Qwen3-0.6B experiment, which failed due to CPU backend limitations, the Qwen3-1.7B experiment progressed further through initialization.

The process was terminated by the operating system before training began.

### Error

```text
Killed
Exit Status: 137
```

### Analysis

Exit code 137 indicates that the process was terminated by the operating system due to memory pressure.

The larger Qwen3-1.7B configuration requires substantially more memory than the 0.6B model because:

- Embedding dimension increased from 1024 → 2048
- MLP dimension increased from 3072 → 6144

As a result, the Colab CPU runtime was unable to allocate sufficient memory for model initialization.

### Conclusion

CPU execution of the scaled model failed due to memory limitations rather than backend compatibility issues.

````python
%%bash
cd /content/maxtext

python3 -m maxtext.trainers.pre_train.train \
src/maxtext/configs/base.yml \
model_name=qwen3-1.7b \
dataset_type=synthetic \
run_name=qwen3_1_7b_gpu \
steps=50 \
per_device_batch_size=1 \
hardware=gpu \
enable_checkpointing=false
````

## GPU Result

The Qwen3-1.7B model initialized successfully and progressed further than the CPU execution.

Execution failed during Transformer Engine initialization.

### Error

```text
OSError:
libcudnn_heuristic.so.9: undefined symbol:
_ZTVN5cudnn7backend12OperationSetE
```

### Analysis

The failure occurred while loading Transformer Engine GPU dependencies.

The error indicates an incompatibility between:

- Transformer Engine
- CUDA runtime
- cuDNN libraries available in the Colab environment

The model configuration itself was accepted successfully.

### Conclusion

GPU execution of the scaled model failed because of a dependency/runtime compatibility issue rather than model configuration problems.

The TPU runtime issue observed during the 0.6B experiment originates from the environment before model compilation begins. Since the failure occurs prior to model execution, the same issue is expected to affect the scaled configuration as well.
